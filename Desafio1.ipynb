### Opci√≥n 2: Datos m√°s realistas

```python
import pandas as pd
import numpy as np
from faker import Faker

faker = Faker('es_CO')
faker.random

departments = []
jobs = []
employees = []

start_date = datetime(2007, 1, 1)
end_date = datetime(2024, 2, 29)

for _ in range(10):
    departments.append(faker.department_code())

for department in departments:
    for _ in range(np.random.randint(2, 5)):
        jobs.append(f"{faker.job()} - {department}")

for job_title in jobs:
    employee = {

        "employee_name": faker.name(),
        "employee_email": faker.email(),
        "phone": faker.phone_number(),
        "job_title": job_title,
        "hire_date": faker.date_between_dates(start_date, end_date)

    }
    employees.append(employee)

departments_df = pd.DataFrame(departments, columns=["department_id"])
jobs_df = pd.DataFrame(jobs, columns=["job_title"])
employees_df = pd.DataFrame(employees)

display(departments_df.head())
display(jobs_df.head())
display(employees_df.head())
```

Se hace uso del archivo csv para cargar la base de datos SQL, es una base de datos SQL Server desplegada en AWS RDS.

```python
import pyodbc
import csv
import os
from dotenv import load_dotenv

class DatabaseLoader:
    def __init__(self, server, port, database, username, password):
        self.server = server
        self.port = port
        self.database = database
        self.username = username
        self.password = password
        self.connection = None
        self.cursor = None

    def connect(self):
        try:
            connection_string = f"DRIVER={{ODBC Driver 17 for SQL Server}};SERVER={self.server},{self.port};DATABASE={self.database};UID={self.username};PWD={self.password}"
            self.connection = pyodbc.connect(connection_string)
            self.cursor = self.connection.cursor()
            print("Connection established successfully.")
        except pyodbc.Error as e:
            print(f"Error connecting to the database: {e}")

    def disconnect(self):
        if self.cursor:
            self.cursor.close()
        if self.connection:
            self.connection.close()
        print("Connection closed successfully.")

    # Desaf√≠o 3
    def load_csv_to_database(self, csv_file, table_name):
        try:
            with open(csv_file, 'r', encoding='utf-8-sig') as file:
                reader = csv.reader(file)
                columns = next(reader)
                query = f"INSERT INTO dbo.{table_name} ({', '.join(columns)}) VALUES ({', '.join(['?' for _ in columns])})"
                print(f"üë©üèª‚Äçüíª SaraDev ==>> query: {query}")
                print(f"Query executed: {query}")
                for row in reader:
                    self.cursor.execute(query, row)
                self.connection.commit()
            print("Data loaded successfully.")
        except FileNotFoundError:
            print("CSV file not found.")
        except pyodbc.Error as e:
            print(f"Error loading data into the database: {e}")

    # Desaf√≠o 4
    def select_from_table(self, table_name):
        try:
            self.cursor.execute(f"SELECT * FROM {table_name}")
            rows = self.cursor.fetchall()
            for row in rows:
                print(row)
        except pyodbc.Error as e:
            print(f"Error executing SELECT query: {e}")

if __name__ == "__main__":
    load_dotenv()
    server = os.environ.get('DB_SERVER')
    port = os.environ.get('DB_PORT')
    database = os.environ.get('DB_NAME')
    username = os.environ.get('DB_USER')
    password = os.environ.get('DB_PASSWORD')

    loader = DatabaseLoader(server, port, database, username, password)
    loader.connect()
    csv_file = os.path.join(os.getcwd(), 'employees_data.csv')
    table_name = 'employees'
    loader.load_csv_to_database(csv_file, table_name)
    loader.select_from_table(table_name)
    loader.disconnect()
```

### DESAFIO #5: Desarrolle una API REST para consultar la view/query/report. Para el desarrollo de la API considere alg√∫n framework de Python, C#/.Net.

### C√≥digo intento aws lambda

```python
import pyodbc
import os
from dotenv import load_dotenv
import json

load_dotenv()
server = os.environ.get('DB_SERVER')
port = os.environ.get('DB_PORT')
database = os.environ.get('DB_NAME')
username = os.environ.get('DB_USER')
password = os.environ.get('DB_PASSWORD')

def get_db_connection():
    try:
        connection_string = f"DRIVER={{ODBC Driver 17 for SQL Server}};SERVER={server},{port};DATABASE={database};UID={username};PWD={password}"
        return pyodbc.connect(connection_string)
    except pyodbc.Error as e:
        print(f"Error connecting to the database: {e}")
        return None

def lambda_handler(event, context):
    if event['httpMethod'] == 'GET' and event['path'] == '/employees':
        connection = get_db_connection()
        if connection:
            with connection:
                with connection.cursor() as cursor:
                    cursor.execute("SELECT * FROM employees")
                    rows = cursor.fetchall()
                    employees = []
                    for row in rows:
                        employee = {
                            'employee_id': row.employee_id,
                            'employee_name': row.employee_name,
                            'employee_email': row.employee_email,
                            'phone': row.phone,
                            'job_title': row.job_title,
                            'hire_date': row.hire_date.strftime('%Y-%m-%d')
                        }
                        employees.append(employee)
                    return {
                        'statusCode': 200,
                        'body': json.dumps(employees),
                        'headers': {
                            'Content-Type': 'application/json'
                        }
                    }
        else:
            return {
                'statusCode': 500,
                'body': json.dumps({'error': 'Database connection error'})
            }
    else:
        return {
            'statusCode': 404,
            'body': json.dumps({'error': 'Endpoint not found'})
        }
```
